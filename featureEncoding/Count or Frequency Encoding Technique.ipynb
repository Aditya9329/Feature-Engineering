 ### Note:
    This encoding technique technique is applied where tyou have higher number of categories.
    This technique basically replace each label of categorical variable by the count
 ### Advantanges
     It is simple to implement
     It does not inccrease the feature dimensional space
 
 ### Disadvantages
      If the some of the lableshas same count , then we will replace with the same count and it will losesome valuable information
